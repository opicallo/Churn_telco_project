```python
dataset = pd.read_excel('dataset_clean.xlsx')
```

### We drop all the variables that are not useful for the model

```python
dataset.drop(['Zip Code','CustomerID', 'Count', 'State', 'Country', 'Lat Long', 'Latitude', 'Longitude','City','Churn Label',], axis=1, inplace=True)
```

### Machine learning algorithms cannot work directly with categorical data and we do need to do some amount of engineering and transformations on this data before we can start modeling on your data. In this case we have converted categorical variables into into dummy variables

```python
Variables = ['Gender', 'Multiple Lines','Senior Citizen','Partner','Dependents','Paperless Billing', 'Internet Service', 'Online Security',
       'Online Backup', 'Device Protection', 'Tech Support', 'Streaming TV',
       'Streaming Movies', 'Payment Method', 'Contract', 'Phone Service']
dataset = pd.get_dummies(dataset, columns=Variables, drop_first=True)

dataset.corr()['Churn Value'].sort_values(ascending = False).plot(kind='bar')
```

### Columns usually have different distributions and scales. This fact may increase the difficulty of the problem being modeled. We used a normalization technique so that all values are within the new range of 0 and 1

```python

sc = MinMaxScaler()
dataset['Tenure Months']=sc.fit_transform(dataset[['Tenure Months']])
dataset['Total Charges']=sc.fit_transform(dataset[['Total Charges']])
dataset['Monthly Charges']=sc.fit_transform(dataset[['Monthly Charges']])
dataset['CLTV']=sc.fit_transform(dataset[['CLTV']])
```
